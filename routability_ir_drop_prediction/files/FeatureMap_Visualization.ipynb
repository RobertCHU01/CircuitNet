```python
import torch
import os
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set the path to the directory where the files are located
save_path = '../work_dir/irdrop_mavi/'
result_dict = {}
```

```python
#IR_drop
# ir = save_path + 'IR_drop/1-RISCY-a-1-c2-u0.7-m1-p1-f0'
for i in range (1, 21):
    file_name = save_path + f'bottleneck_output_{i*40}.pth'

    # Load the file
    data = torch.load(file_name)

    # Assuming the data is already a tensor
    tensor = data
    # Reshape the tensor to 2D format
    tensor = tensor.reshape(-1, 512 * 8 * 32 * 32)  # (-1, flattened_features)
    # Convert the tensor to a NumPy array
    tensor_np = tensor.detach().numpy()

    # Visualize the 2D tensor using seaborn.heatmap
    fig, ax = plt.subplots(figsize=(10, 6))
    sns.heatmap(tensor_np, annot=False, cmap='viridis', ax=ax)
    plt.show()
    # print(tensor_data.shape)
```
