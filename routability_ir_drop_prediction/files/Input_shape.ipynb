```python
def countlarger(tensor):
    """
    Counts the number of occurrences of the value 1 in a PyTorch tensor.
    
    Args:
        tensor (torch.Tensor): The input tensor.
        
    Returns:
        int: The number of occurrences of the value 1 in the tensor.
    """
    return torch.sum(tensor > 1).item()
def count_ones(tensor):
    """
    Counts the number of occurrences of the value 1 in a PyTorch tensor.
    
    Args:
        tensor (torch.Tensor): The input tensor.
        
    Returns:
        int: The number of occurrences of the value 1 in the tensor.
    """
    return torch.sum(tensor == 1).item()

def count_zeros(tensor):
    """
    Counts the number of occurrences of the value 1 in a PyTorch tensor.
    
    Args:
        tensor (torch.Tensor): The input tensor.
        
    Returns:
        int: The number of occurrences of the value 1 in the tensor.
    """
    return torch.sum(tensor == 0).item()

def print_non_zeros(tensor):
    """
    Prints all the non-zero elements in a PyTorch tensor.

    Args:
        tensor (torch.Tensor): The input tensor.
    """
    non_zeros = tensor[tensor != 0]
    print("Non-zero elements in the tensor:")
    print(non_zeros)
```
